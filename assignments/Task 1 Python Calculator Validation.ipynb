# Task 1: Python Calculator Validation

```python
def sum(num_1, num_2):
    return num_1 + num_2

def sub(num_1, num_2):
    return num_1 - num_2

def div(num_1, num_2):
    return num_1 / num_2

def multply(num_1, num_2):
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    try:
        choice = input("Enter the operation (1/2/3/4): ")
        
        # validation choice
        if choice not in ["1", "2", "3", "4"]:
            print("Error: Invalid choice! Please selec a number from 1 to 4.")
            return

        num_1 = float(input("Enter first number: "))
        num_2 = float(input("Enter second number: "))
        
        result = 0
        
        if choice == "1":
            result = sum(num_1, num_2)
        elif choice == "2":
            result = sub(num_1, num_2)
        elif choice == "3":
            result = div(num_1, num_2)
        elif choice == "4":
            result = multply(num_1, num_2)
            
        print("Result: ", result)

    except ValueError:
        # check numeric values
        print("Error: Invalid input! Please enter numeric values only.")
        
    except ZeroDivisionError:
        # can't divided by zero
        print("Error: Cannot divide by zero.")
        
    except Exception as e:
        # handling any unexpected errors
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
